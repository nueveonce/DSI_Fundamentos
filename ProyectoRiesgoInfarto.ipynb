<h1 style="color: rgb(173, 193, 94); font-weight: bold">üìó Correlacion RPM (Recursive Feature Elimination)</h1>
<p>RFE es una t√©cnica de selecci√≥n de caracter√≠sticas que:

Entrena el modelo usando todas las caracter√≠sticas
Asigna una importancia a cada caracter√≠stica
Elimina la caracter√≠stica menos importante
Repite el proceso hasta alcanzar el n√∫mero deseado de caracter√≠sticas</p>

<h1 style="color: rgb(173, 193, 94); font-weight: bold">üìó SelectKBest</h1>
<p>SelectKBest
Es una t√©cnica de selecci√≥n de caracter√≠sticas que:

Calcula una puntuaci√≥n para cada caracter√≠stica usando diferentes m√©tricas estad√≠sticas (chi-cuadrado, ANOVA F-value, etc.)
Selecciona las k mejores caracter√≠sticas basadas en esas puntuaciones</p>
